```python
# nearest mean
def nrst_to_mean(li):
    me=np.mean(li)
    nres=min(range(len(li)),key=lambda x:abs(li[x]-me))
    return li[nres]
```

```python
# Ticket price where origin is not equal as destination
# dict for every airline ,aircraft,distance,price




AADT = {
    i: {
        m: {
            q: nrst_to_mean(
                df[
                    (df["Airline"] == i) &
                    (df["Aircraft_Type"] == m) &
                    (df["Distance_km"] == q) &
                    (df["Ticket_Price"].notna())
                ]["Ticket_Price"].tolist()
            )
            for q in n
        }
        for m, n in {
            k: df[
                (df["Airline"] == i) &
                (df["Aircraft_Type"] == k) &
                (df["Distance_km"] != 0.0)
            ]["Distance_km"].unique().tolist()
            for k in j
        }.items()
    }
    for i, j in {
        i: df[df["Airline"] == i]["Aircraft_Type"].unique()
        for i in df["Airline"].unique()
    }.items()
}


for i, j in AADT.items():
    for m, n in j.items():
        for q, w in n.items():
            df.loc[
                (df["Airline"] == i) &
                (df["Aircraft_Type"] == m) &
                (df["Distance_km"] == q) &
                (df["Ticket_Price"].isna()),
                "Ticket_Price"
            ] = w



```

```python
AADWS = {
    i: {
        j: {
            d: {
                w: (
                    df[
                        (df["Airline"] == i) &
                        (df["Aircraft_Type"] == j) &
                        (df["Distance_km"] == d) &
                        (df["Weather_Condition"] == w) &
                        (df["Duration_hours"].notna())
                    ]["Distance_km"]
                    /
                    df[
                        (df["Airline"] == i) &
                        (df["Aircraft_Type"] == j) &
                        (df["Distance_km"] == d) &
                        (df["Weather_Condition"] == w) &
                        (df["Duration_hours"].notna())
                    ]["Duration_hours"]
                ).to_list()
                for w in q
            }
            for d, q in df[
                (df["Airline"] == i) &
                (df["Aircraft_Type"] == j) &
                (df["Distance_km"] != 0.0)
            ].groupby("Distance_km")["Weather_Condition"]
             .unique().to_dict().items()
        }
        for j in df[df["Airline"] == i]
        .groupby("Aircraft_Type")["Distance_km"]
        .unique().to_dict()
    }
    for i in df.groupby("Airline")["Aircraft_Type"]
    .unique().to_dict()
}
```

```python
def dhf(a):
    h=int(a)
    m=int((a-h)*60)
    s=round(((a-h)*60 -m)*60)
    if s>=60:
        m+=s//60
        s=s%60
    if m>=60:
        h+=m//60
        m=m%60
    return f"{h:02d}:{m:02d}:{s:02d}"
df["Duration_hours"]=df["Duration_hours"].apply(lambda x:dhf(x))
```

```python
def dmf(a):
    h=0
    m=int(a)
    s=round((a-m)*60)
    if s>=60:
        m+=s//60
        s=s%60
    if m>=60:
        h+=m//60
        m=m%60
    return f"{h:02d}:{m:02d}:{s:02d}"
df["Delay_minutes"]=df["Delay_minutes"].apply(lambda x:dhf(x))
```
